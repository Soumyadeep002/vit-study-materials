# **Integration**

```python
from scipy.integrate import quad, dblquad, tplquad
import numpy as np
```

### **Integration**

```python
fx = lambda x: np.exp(-x**2)

lx = 0
ux = 4

val, err = quad(fx, lx, ux)
print(f'Value: {val}')
print(f'Error: {err}')
```

### **Double Integration**

```python
fx = lambda y, x: (x**2)+y

lx = 0
ux = 2

ly = lambda x: 0
uy = lambda x: x

val, err = dblquad(fx, lx, ux, ly, uy)
print(f'Value: {val}')
print(f'Error: {err}')
```

### **Triple integration**

```python
fx = lambda y,z,x : (x*y**2) + z 
ly = lambda x, z: z
uy = lambda x, z: (1- 2*x)+z

lz = lambda x: -x
uz = lambda x: x

lx = 0
ux = 0.5

val, err = tplquad(fx, lx, ux, lz, uz, ly, uy)
print(f'Value: {val}')
print(f'Error: {err}')
```

# **Interpolation**

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 10)
y = 2*x**2 + 3*x- 5
y
```

```python
lp = interp1d(x,y, kind='linear')

fig, axs = plt.subplots(figsize = (12,6))
axs.plot(x,y, 'or')
axs.plot(x, lp(x), '-b')
```

```python
def allOddDigits(n):
    for i in list(str(n)):
        if int(i)%2 == 0:
            msg = 'There is a even number it has'
            break
    else:
        msg = 'All are ODD digits'
        
    return msg

allOddDigits(347)
```

# **Optimization**

```python
from scipy.optimize import curve_fit

x = np.linspace(0,8,10)
y= np.array([0.48077, 0.86895, 0.97010, 0.70940, 0.27965, -0.00138, 0.07410, 0.45031, 0.84903, 0.97545])
fx = lambda x,a,b,c:a*np.sin(b*x) + c


fig, axs = plt.subplots()
axs.plot(x, y, 'or')
plt.show
```

```python
popt, pcov = curve_fit(fx, x, y)
popt
```

```python
def fxn(x, *popt):
    return popt[0]*np.sin(popt[1]*x) + popt[2]
```

```python
xn = np.linspace(0,8, 100)

fig, axs = plt.subplots()
axs.plot(x, y, 'or')
axs.plot(xn, fxn(xn, *popt), '-b')
plt.show
```

# **Linear Algebra**

```python
from scipy.linalg import solve
arr1 = np.array([[1,2,5],[2,-5,1],[2,-3,8]])
arr2 = np.array([9,8,2])
sol = solve(arr1, arr2)
sol
```

```python
from scipy.linalg import det, eig

val, vec = eig(arr1)
print(det(arr1))
print(f'Eigen Value: {val}')
print(f'Eigen Vectors: {vec}')
```

```python
from scipy.linalg import svd

u, s, vh = svd(arr1)
print(f'U = {u}')
print(f'S = {s}')
print(f'VH = {vh}')
```
