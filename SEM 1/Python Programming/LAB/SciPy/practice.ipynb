```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.integrate import quad, tplquad
fx = lambda x:np.exp(-x**2)
val, err = quad(fx, 0, 4)

val
```

```python
fx = lambda y,z,x: (x*y**2) + z

lz = lambda x: -x
uz = lambda x: x

ly = lambda x,z: z
uy = lambda x,z: 1 - 2*x+z

val, err = tplquad(fx, 0, 0.5, lz, uz, ly, uy)

val
```

```python
from scipy.optimize import curve_fit
from scipy.interpolate import interp1d
x = np.linspace(0,8,10)
xrange = np.linspace(0,8,100)
y= np.array([0.48077, 0.86895, 0.97010, 0.70940, 0.27965, -0.00138, 0.07410, 0.45031, 0.84903, 0.97545])


fx = lambda x,a,b,c:a*np.sin(b*x) + c

lp = interp1d(x, y, kind='cubic')

fig, axs = plt.subplots()
axs.plot(x,y, 'o')
axs.plot(xrange,lp(xrange), '--r', lw=1)


popt, pcuv = curve_fit(fx, x, y)
popt

```

```python
def func(x, *popt):
    return popt[0]*np.sin(popt[1]*x) + popt[2]
```

```python
lp1 =interp1d(x, func(x, *popt), kind='cubic')

fig, axs = plt.subplots()
axs.plot(x,func(x, *popt), 'o')
axs.plot(xrange,lp1(xrange), '--r', lw=1)
```
